```python
from qiskit import QuantumCircuit
# Qubit index:

#0  → fermion 0
#1  → fermion 1

#2  → boson mode 0, bit 0 (LSB)
#3  → boson mode 0, bit 1
#4  → boson mode 0, bit 2
#5  → boson mode 0, bit 3 (MSB)

#6–9   → boson mode 1
##14–17 → boson mode 3
qc = QuantumCircuit(18)
```

```python
import numpy as np
### Fermion part of Hamiltonian in eq. 12 of 2509.11477
#Define dt
dt = 1.0
#Define m_\psi and n
mpsi=1
b=1
# Add Trotter loop

for _ in range(5):

     # --- mψ Z₁ term ---
    qc.rz(2 * mpsi * dt, 1)

    # --- (1/2b) X₁ term ---
    qc.rx(2 * (dt/(2*b)), 1)

    # --- (1/2b) X₀ X₁ term ---
    theta = dt/(2*b)
    # HXH=Z, X=HZH, X0​X1​=(H0​Z0​H0​)(H1​Z1​H1​)

    #Apply H gate on both qubit 
    qc.h(0)
    qc.h(1)
    #CNOT maps Z_1--> Z_0Z_1, then apply Z gate
    qc.cx(0, 1)
    qc.rz(2 * theta, 1)
    qc.cx(0, 1)
    # Reverse h gate
    qc.h(0)
    qc.h(1)
```
